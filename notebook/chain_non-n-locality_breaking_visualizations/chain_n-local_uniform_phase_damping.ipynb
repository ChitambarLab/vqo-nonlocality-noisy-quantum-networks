```python
from context import qnetvo as qnet
from context import utilities

import matplotlib.pyplot as plt
from pennylane import numpy as np
```

```python
data_dir = "../../script/data/chain_n-local_uniform_phase_damping/"

n3_max_ent_local_rot_files = utilities.get_data_files(data_dir, r"max_entangled_local_rot_n-3_.*")
n3_max_ent_arb_files = utilities.get_data_files(data_dir, r"max_entangled_arb_n-3_.*")
n3_arb_local_rot_files = utilities.get_data_files(data_dir, r"arb_local_rot_n-3_.*")
n3_arb_arb_files = utilities.get_data_files(data_dir, r"arb_arb_n-3_.*")

n4_max_ent_local_rot_files = utilities.get_data_files(data_dir, r"max_entangled_local_rot_n-4_.*")
n4_max_ent_arb_files = utilities.get_data_files(data_dir, r"max_entangled_arb_n-4_.*")
n4_arb_local_rot_files = utilities.get_data_files(data_dir, r"arb_local_rot_n-4_.*")
n4_arb_arb_files = utilities.get_data_files(data_dir, r"arb_arb_n-4_.*")

print("num files : ")

print("\tn3 max_ent_local_rot : ", len(n3_max_ent_local_rot_files))
print("\tn3 max_ent_ghz_rot : ", len(n3_max_ent_arb_files))
print("\tn3 arb_local_rot : ", len(n3_arb_local_rot_files))
print("\tn3 arb_ghz_rot : ", len(n3_arb_arb_files))

print("\tn4 max_ent_local_rot : ", len(n4_max_ent_local_rot_files))
print("\tn4 max_ent_ghz_rot : ", len(n4_max_ent_arb_files))
print("\tn4 arb_local_rot : ", len(n4_arb_local_rot_files))
print("\tn4 arb_ghz_rot : ", len(n4_arb_arb_files))
```

```python
n3_max_ent_local_rot = utilities.analyze_data_one_param_scan(n3_max_ent_local_rot_files)
n3_max_ent_arb = utilities.analyze_data_one_param_scan(n3_max_ent_arb_files)
n3_arb_local_rot = utilities.analyze_data_one_param_scan(n3_arb_local_rot_files)
n3_arb_arb = utilities.analyze_data_one_param_scan(n3_arb_arb_files)

n4_max_ent_local_rot = utilities.analyze_data_one_param_scan(n4_max_ent_local_rot_files)
n4_max_ent_arb = utilities.analyze_data_one_param_scan(n4_max_ent_arb_files)
n4_arb_local_rot = utilities.analyze_data_one_param_scan(n4_arb_local_rot_files)
n4_arb_arb = utilities.analyze_data_one_param_scan(n4_arb_arb_files)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("n-Local Chain Network with Uniform Amplitude Damping", fontsize=18)

ax1.set_title("3-local Chain", size=16)
ax1.set_ylabel("3-Local Chain 22 Score", size=14)
ax1.set_xlabel("Noise Parameter", size=14)

ax1.plot(n3_arb_arb["noise_params"], [np.sqrt(2)]*len(n3_arb_arb["noise_params"]), label="Quantum Bound")
ax1.plot(n3_arb_arb["noise_params"], [1]*len(n3_arb_arb["noise_params"]), label="Classical Bound")

ax1.plot(n3_max_ent_local_rot["noise_params"], n3_max_ent_local_rot["max_scores"], label="Max Ent Local Rot")
ax1.plot(n3_max_ent_arb["noise_params"], n3_max_ent_arb["max_scores"], label="Max Ent Arb Meas")
ax1.plot(n3_arb_local_rot["noise_params"], n3_arb_local_rot["max_scores"], label="Arb Local Rot")
ax1.plot(n3_arb_arb["noise_params"], n3_arb_arb["max_scores"], label="Arb Prep Arb Meas")

ax1.legend()


ax2.set_title("4-local Chain", size=16)
ax2.set_ylabel("4-Local Chain 22 Score", size=14)
ax2.set_xlabel("Noise Parameter", size=14)

ax2.plot(n4_arb_arb["noise_params"], [np.sqrt(2)]*len(n4_arb_arb["noise_params"]), label="Quantum Bound")
ax2.plot(n4_arb_arb["noise_params"], [1]*len(n4_arb_arb["noise_params"]), label="Classical Bound")

ax2.plot(n4_max_ent_local_rot["noise_params"], n4_max_ent_local_rot["max_scores"], label="Max Ent Local Rot")
ax2.plot(n4_max_ent_arb["noise_params"], n4_max_ent_arb["max_scores"], label="Max Ent Arb Meas")
ax2.plot(n4_arb_local_rot["noise_params"], n4_arb_local_rot["max_scores"], label="Arb Local Rot")
ax2.plot(n4_arb_arb["noise_params"], n4_arb_arb["max_scores"], label="Arb prep Arb meas")

ax2.legend()


plt.show()
```
