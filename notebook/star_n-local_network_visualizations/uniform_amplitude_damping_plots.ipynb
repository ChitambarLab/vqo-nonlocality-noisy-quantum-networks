```python
import qnetvo as qnet
from context import utilities

import matplotlib.pyplot as plt
from pennylane import numpy as np
```

```python
data_dir = "../../script/data/star_n-local_uniform_amplitude_damping/"

n3_max_ent_local_rot_files = utilities.get_data_files(data_dir, r"max_entangled_local_rot_n-3_.*")
n3_max_ent_ghz_rot_files = utilities.get_data_files(data_dir, r"max_entangled_ghz_rot_n-3_.*")
n3_arb_local_rot_files = utilities.get_data_files(data_dir, r"arb_local_rot_n-3_.*")
n3_arb_ghz_rot_files = utilities.get_data_files(data_dir, r"arb_ghz_rot_n-3_.*")

n4_max_ent_local_rot_files = utilities.get_data_files(data_dir, r"max_entangled_local_rot_n-4_.*")
n4_max_ent_ghz_rot_files = utilities.get_data_files(data_dir, r"max_entangled_ghz_rot_n-4_.*")
n4_arb_local_rot_files = utilities.get_data_files(data_dir, r"arb_local_rot_n-4_.*")
n4_arb_ghz_rot_files = utilities.get_data_files(data_dir, r"arb_ghz_rot_n-4_.*")

print("num files : ")

print("\tn3 max_ent_local_rot : ", len(n3_max_ent_local_rot_files))
print("\tn3 max_ent_ghz_rot : ", len(n3_max_ent_ghz_rot_files))
print("\tn3 arb_local_rot : ", len(n3_arb_local_rot_files))
print("\tn3 arb_ghz_rot : ", len(n3_arb_ghz_rot_files))

print("\tn4 max_ent_local_rot : ", len(n4_max_ent_local_rot_files))
print("\tn4 max_ent_ghz_rot : ", len(n4_max_ent_ghz_rot_files))
print("\tn4 arb_local_rot : ", len(n4_arb_local_rot_files))
print("\tn4 arb_ghz_rot : ", len(n4_arb_ghz_rot_files))
```

```python
n3_max_ent_local_rot = utilities.analyze_data_one_param_scan(n3_max_ent_local_rot_files)
n3_max_ent_ghz_rot = utilities.analyze_data_one_param_scan(n3_max_ent_ghz_rot_files)
n3_arb_local_rot = utilities.analyze_data_one_param_scan(n3_arb_local_rot_files)
n3_arb_ghz_rot = utilities.analyze_data_one_param_scan(n3_arb_ghz_rot_files)

n4_max_ent_local_rot = utilities.analyze_data_one_param_scan(n4_max_ent_local_rot_files)
n4_max_ent_ghz_rot = utilities.analyze_data_one_param_scan(n4_max_ent_ghz_rot_files)
n4_arb_local_rot = utilities.analyze_data_one_param_scan(n4_arb_local_rot_files)
n4_arb_ghz_rot = utilities.analyze_data_one_param_scan(n4_arb_ghz_rot_files)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("n-Local Star Network with Uniform Amplitude Damping", fontsize=18)

ax1.set_title("3-local Star", size=16)
ax1.set_ylabel("Trilocal Star 22 Score", size=14)
ax1.set_xlabel("Noise Parameter", size=14)

ax1.plot(n3_arb_ghz_rot["noise_params"], [np.sqrt(2)]*len(n3_arb_ghz_rot["noise_params"]), label="Quantum Bound")
ax1.plot(n3_arb_ghz_rot["noise_params"], [1]*len(n3_arb_ghz_rot["noise_params"]), label="Classical Bound")

ax1.plot(n3_max_ent_local_rot["noise_params"], n3_max_ent_local_rot["max_scores"], label="Max Ent Local Rot")
ax1.plot(n3_max_ent_ghz_rot["noise_params"], n3_max_ent_ghz_rot["max_scores"], label="Max Ent GHZ Rot")
ax1.plot(n3_arb_local_rot["noise_params"], n3_arb_local_rot["max_scores"], label="Arb Local Rot")
ax1.plot(n3_arb_ghz_rot["noise_params"], n3_arb_ghz_rot["max_scores"], label="Arb Ghz Rot")

ax1.legend()


ax2.set_title("4-local Star", size=16)
ax2.set_ylabel("Quadlocal Star 22 Score", size=14)
ax2.set_xlabel("Noise Parameter", size=14)

ax2.plot(n4_arb_ghz_rot["noise_params"], [np.sqrt(2)]*len(n4_arb_ghz_rot["noise_params"]), label="Quantum Bound")
ax2.plot(n4_arb_ghz_rot["noise_params"], [1]*len(n4_arb_ghz_rot["noise_params"]), label="Classical Bound")

ax2.plot(n4_max_ent_local_rot["noise_params"], n4_max_ent_local_rot["max_scores"], label="Max Ent Local Rot")
ax2.plot(n4_max_ent_ghz_rot["noise_params"], n4_max_ent_ghz_rot["max_scores"], label="Max Ent GHZ Rot")
ax2.plot(n4_arb_local_rot["noise_params"], n4_arb_local_rot["max_scores"], label="Arb Local Rot")
ax2.plot(n4_arb_ghz_rot["noise_params"], n4_arb_ghz_rot["max_scores"], label="Arb Ghz Rot")

ax2.legend()


plt.show()
```
