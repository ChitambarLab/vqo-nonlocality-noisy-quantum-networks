## Loading Data for Amplitude Damping Channel

```python
amp_in_local_ry_files = utilities.get_data_files(
    "../../script/data/star_n-local_1-qubit_amplitude_damping/inside/", r".*_local_ry_n-3_.*"
)
print("num in local_ry files : ", len(amp_in_local_ry_files))

amp_out_local_ry_files = utilities.get_data_files(
    "../../script/data/star_n-local_1-qubit_amplitude_damping/outside/", r".*_local_ry_n-3_.*"
)
print("num out local_ry files : ", len(amp_out_local_ry_files))

amp_in_local_rot_files = utilities.get_data_files(
    "../../script/data/star_n-local_1-qubit_amplitude_damping/inside/", r".*_local_rot_n-3_.*"
)
print("num in local_rot files : ", len(amp_in_local_rot_files))

amp_out_local_rot_files = utilities.get_data_files(
    "../../script/data/star_n-local_1-qubit_amplitude_damping/outside/", r".*_local_rot_n-3_.*"
)
print("num out local_rot files : ", len(amp_out_local_rot_files))

amp_in_ghz_rot_files = utilities.get_data_files(
    "../../script/data/star_n-local_1-qubit_amplitude_damping/inside/", r".*_ghz_rot_n-3_.*"
)
print("num in ghz_rot files : ", len(amp_in_ghz_rot_files))

amp_out_ghz_rot_files = utilities.get_data_files(
    "../../script/data/star_n-local_1-qubit_amplitude_damping/outside/", r".*_ghz_rot_n-3_.*"
)
print("num out ghz_rot files : ", len(amp_out_ghz_rot_files))
```

## Analyzing Depolarizing Channel Data

```python
dep_in_local_ry = utilities.analyze_data_one_param_scan(dep_in_local_ry_files)
dep_out_local_ry = utilities.analyze_data_one_param_scan(dep_out_local_ry_files)
dep_in_local_rot = utilities.analyze_data_one_param_scan(dep_in_local_rot_files)
dep_out_local_rot = utilities.analyze_data_one_param_scan(dep_out_local_rot_files)
dep_in_ghz_rot = utilities.analyze_data_one_param_scan(dep_in_ghz_rot_files)
dep_out_ghz_rot = utilities.analyze_data_one_param_scan(dep_out_ghz_rot_files)

dep_local_ry_max_scores = list(map(lambda x,y: max(x,y), dep_in_local_ry["max_scores"], dep_out_local_ry["max_scores"]))
dep_local_rot_max_scores = list(map(lambda x,y: max(x,y), dep_in_local_rot["max_scores"], dep_out_local_rot["max_scores"]))
dep_ghz_rot_max_scores = list(map(lambda x,y: max(x,y), dep_in_ghz_rot["max_scores"], dep_out_ghz_rot["max_scores"]))
```

## Analyzing Amplitude Damping Channel

```python
amp_in_local_ry = utilities.analyze_data_one_param_scan(amp_in_local_ry_files)
amp_out_local_ry = utilities.analyze_data_one_param_scan(amp_out_local_ry_files)
amp_in_local_rot = utilities.analyze_data_one_param_scan(amp_in_local_rot_files)
amp_out_local_rot = utilities.analyze_data_one_param_scan(amp_out_local_rot_files)
amp_in_ghz_rot = utilities.analyze_data_one_param_scan(amp_in_ghz_rot_files)
amp_out_ghz_rot = utilities.analyze_data_one_param_scan(amp_out_ghz_rot_files)

amp_local_ry_max_scores = list(map(lambda x,y: max(x,y), amp_in_local_ry["max_scores"], amp_out_local_ry["max_scores"]))
amp_local_rot_max_scores = list(map(lambda x,y: max(x,y), amp_in_local_rot["max_scores"], amp_out_local_rot["max_scores"]))
amp_ghz_rot_max_scores = list(map(lambda x,y: max(x,y), amp_in_ghz_rot["max_scores"], amp_out_ghz_rot["max_scores"]))
```

## Plotting Data

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Trilocal Star with Bell State Preparation and one Noisy Qubit", fontsize=18)

ax1.set_title("Depolarizing Noise", size=16)
ax1.set_ylabel("Trilocal Star 22 Score", size=14)
ax1.set_xlabel("Noise Parameter", size=14)

ax1.plot(dep_in_local_ry["noise_params"], [np.sqrt(2)]*len(dep_in_local_ry["noise_params"]), label="Quantum Bound")
ax1.plot(dep_in_local_ry["noise_params"], [1]*len(dep_in_local_ry["noise_params"]), "--", label="Local Bound")

ax1.plot(dep_in_local_ry["noise_params"], dep_local_ry_max_scores, ":d", markersize=8, label="Local RY Measurements")
ax1.plot(dep_in_local_rot["noise_params"], dep_local_rot_max_scores, ":^", markersize=8, label="Local Rot Measurements")
ax1.plot(dep_in_ghz_rot["noise_params"], dep_ghz_rot_max_scores, ":.", markersize=10, label="GHZ Rot Measurements")

ax1.plot([(1-1/np.sqrt(2**3))*3/4], [1], "b*", markersize=8, label="Noise Threshold")

ax1.legend()

ax2.set_title("Amplitude Damping Noise", size=16)
ax2.set_ylabel("Trilocal Star 22 Score", size=14)
ax2.set_xlabel("Noise Parameter", size=14)

ax2.plot(amp_out_local_ry["noise_params"], [np.sqrt(2)]*len(amp_out_local_ry["noise_params"]), label="Quantum Bound")
ax2.plot(amp_out_local_ry["noise_params"], [1]*len(amp_out_local_ry["noise_params"]), "--", label="Local Bound")

ax2.plot(amp_out_local_ry["noise_params"], amp_local_ry_max_scores, ":d", markersize=8, label="Local RY Measurements")
ax2.plot(amp_out_local_rot["noise_params"], amp_local_rot_max_scores, ":^", markersize=8, label="Local Rot Measurements")
ax2.plot(amp_out_ghz_rot["noise_params"], amp_ghz_rot_max_scores, ":.", markersize=10, label="GHZ Measurements")

ax2.plot([1-1/(2**3)], [1], "b*", markersize=8, label="Noise Threshold")

ax2.legend()


plt.show()
```
