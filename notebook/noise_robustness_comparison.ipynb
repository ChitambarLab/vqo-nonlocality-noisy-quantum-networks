```python
fig, (
    (ax5, ax3, ax1), (ax6, ax4, ax2) 
) = plt.subplots(2, 3, figsize=(20,15))

fig.suptitle("VQO of Non-n-Locality in Noisy Networks", fontsize=18)

num_samples = 21
noise_params = np.arange(0,1.001,0.05)
quantum_bound = [np.sqrt(2)] * num_samples
classical_bound = [1] * num_samples

ax1.set_title("Single-Qubit Phase Damping", size=16)
ax1.set_ylabel("Bell Score", size=14)
ax1.set_xlabel("Noise Parameter", size=14)

ax1.plot(noise_params, quantum_bound, label="Quantum Bound")
ax1.plot(noise_params, classical_bound, label="Classical Bound")

ax1.plot(noise_params, max_chsh_single_pd, ":d", label="CHSH")
ax1.plot(noise_params, max_bilocal_single_pd, ":d", label="Bilocal")
ax1.plot(noise_params, max_n3_chain_single_pd,":.", label="3-Local Chain")
ax1.plot(noise_params, max_n4_chain_single_pd,":d", label="4-Local Chain")
ax1.plot(noise_params, max_n3_star_single_pd,":^", label="3-Local Star")
ax1.plot(noise_params, max_n4_star_single_pd, ":x",label="4-Local Star")


ax2.set_title("Uniform Phase Damping", size=16)
ax2.set_ylabel("Bell Score", size=14)
ax2.set_xlabel("Noise Parameter", size=14)

ax2.plot(noise_params, quantum_bound, label="Quantum Bound")
ax2.plot(noise_params, classical_bound, label="Classical Bound")

ax2.plot(noise_params, max_chsh_uniform_pd, ":d", label="CHSH")
ax2.plot(noise_params, max_bilocal_uniform_pd, ":d", label="Bilocal")
ax2.plot(noise_params, max_n3_chain_uniform_pd,":.", label="3-Local Chain")
ax2.plot(noise_params, max_n4_chain_uniform_pd,":d", label="4-Local Chain")
ax2.plot(noise_params, max_n3_star_uniform_pd,":^", label="3-Local Star")
ax2.plot(noise_params, max_n4_star_uniform_pd, ":x",label="4-Local Star")

ax2.legend()

ax3.set_title("Single-Qubit Amplitude Damping", size=16)
ax3.set_ylabel("Bell Score", size=14)
ax3.set_xlabel("Noise Parameter", size=14)

ax3.plot(noise_params, quantum_bound, label="Quantum Bound")
ax3.plot(noise_params, classical_bound, label="Classical Bound")

ax3.plot(noise_params, max_chsh_single_ad, ":d", label="CHSH")
ax3.plot(noise_params, max_bilocal_single_ad, ":d", label="Bilocal")

ax3.plot(noise_params, max_n3_chain_single_ad,":.", label="3-Local Chain")
ax3.plot(noise_params, max_n4_chain_single_ad,":d", label="4-Local Chain")
ax3.plot(noise_params, max_n3_star_single_ad,":^", label="3-Local Star")
ax3.plot(noise_params, max_n4_star_single_ad, ":x",label="4-Local Star")

ax4.set_title("Uniform Amplitude Damping", size=16)
ax4.set_ylabel("Bell Score", size=14)
ax4.set_xlabel("Noise Parameter", size=14)

ax4.plot(noise_params, quantum_bound, label="Quantum Bound")
ax4.plot(noise_params, classical_bound, label="Classical Bound")

ax4.plot(noise_params, max_chsh_uniform_ad, ":d", label="CHSH")
ax4.plot(noise_params, max_bilocal_uniform_ad, ":d", label="Bilocal")
ax4.plot(noise_params, max_n3_chain_uniform_ad, ":.", label="3-Local Chain")
ax4.plot(noise_params, max_n4_chain_uniform_ad, ":d", label="4-Local Chain")
ax4.plot(noise_params, max_n3_star_uniform_ad, ":^", label="3-Local Star")
ax4.plot(noise_params, max_n4_star_uniform_ad, ":x", label="4-Local Star")

ax6.set_title("Uniform Qubit Depolarizing", size=16)
ax6.set_ylabel(r"Bell Score ($S_{Star}$)", size=14)
ax6.set_xlabel("Noise Parameter", size=14)

ax6.plot(noise_params, quantum_bound, label="Quantum Bound")
ax6.plot(noise_params, classical_bound, label="Classical Bound")

ax6.plot(noise_params, max_chsh_uniform_dep, ":d", label="CHSH")
ax6.plot(noise_params, max_bilocal_uniform_dep, ":d", label="Bilocal")
ax6.plot(noise_params, max_n3_chain_uniform_dep, ":.", label="3-Local Chain")
ax6.plot(noise_params, max_n4_chain_uniform_dep, ":d", label="4-Local Chain")
ax6.plot(noise_params, max_n3_star_uniform_dep, ":^", label="3-Local Star")
ax6.plot(noise_params, max_n4_star_uniform_dep, ":x", label="4-Local Star")

ax6.legend()

plt.show()
```
