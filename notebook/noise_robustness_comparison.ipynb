# Plotting Data

```python
fig, (
    (ax1, ax2, ax3), (ax4, ax5, ax6) 
) = plt.subplots(2, 3, figsize=(20,12))

fig.suptitle("VQO of Non-n-Locality on Classical Simulator with Qubit Noise", fontsize=36, fontweight="bold")

num_samples = 21
noise_params = np.arange(0,1.001,0.05)
quantum_bound = [np.sqrt(2)] * num_samples
classical_bound = [1] * num_samples

axes = [ax1, ax2, ax3, ax4, ax5, ax6]

single_dep = [
    max_chsh_single_dep, max_bilocal_single_dep, max_n3_chain_single_dep,
    max_n4_chain_single_dep, max_n3_star_single_dep, max_n4_star_single_dep
]

single_ad = [
    max_chsh_single_ad, max_bilocal_single_ad, max_n3_chain_single_ad,
    max_n4_chain_single_ad, max_n3_star_single_ad, max_n4_star_single_ad
]

single_pd = [
    max_chsh_single_pd, max_bilocal_single_pd, max_n3_chain_single_pd,
    max_n4_chain_single_pd, max_n3_star_single_pd, max_n4_star_single_pd
]

uniform_dep = [
    max_chsh_uniform_dep, max_bilocal_uniform_dep, max_n3_chain_uniform_dep,
    max_n4_chain_uniform_dep, max_n3_star_uniform_dep, max_n4_star_uniform_dep
]

uniform_ad = [
    max_chsh_uniform_ad, max_bilocal_uniform_ad, max_n3_chain_uniform_ad,
    max_n4_chain_uniform_ad, max_n3_star_uniform_ad, max_n4_star_uniform_ad
]

uniform_pd = [
    max_chsh_uniform_pd, max_bilocal_uniform_pd, max_n3_chain_uniform_pd,
    max_n4_chain_uniform_pd, max_n3_star_uniform_pd, max_n4_star_uniform_pd
]

col_titles = ["Depolarizing", "Amplitude Damping", "Phase Damping"]
data_sets = [single_dep, single_ad, single_pd, uniform_dep, uniform_ad, uniform_pd]
row_labels = ["Single Qubit Noise", "Uniform Noise"]


for i in range(6):
    ax = axes[i]

    if i in [0,3]:
        ax.set_ylabel(r"Bell Score ($S_{Star}^n$)", size=28)
        ax.annotate(row_labels[i%2],
            xy=(-1, 0.5),
            xytext=(-ax.yaxis.labelpad,0),                    
            xycoords=ax.yaxis.label,
            textcoords='offset points',
            size=28,
            ha='center',
            va='center',
            rotation=90,
            fontweight="bold",
        )
    if i in [3,4,5]:
        ax.set_xlabel(r"Noise Parameter ($\gamma$)", size=28)
    if i in [0,1,2]:
        ax.set_title(col_titles[i], size=32)
    
    data = data_sets[i]
    
    ax.plot(noise_params, quantum_bound, linewidth=3, label="Quantum Bound")
    ax.plot(noise_params, classical_bound, "--", linewidth=3, label="Classical Bound")
    
    ax.plot(noise_params, data[0], ":s", linewidth=2, markersize=10, label="CHSH")
    ax.plot(noise_params, data[1], ":o", linewidth=2, markersize=8, label="Bilocal")
    ax.plot(noise_params, data[2], ":^", linewidth=2, markersize=9, label="3-Local Chain")
    ax.plot(noise_params, data[3], ":d", linewidth=2, markersize=9, label="4-Local Chain")
    ax.plot(noise_params, data[4], ":*", linewidth=2, markersize=9, label="3-Local Star")
    ax.plot(noise_params, data[5], ":P", linewidth=2, markersize=9, label="4-Local Star")
    
    if i == 0:
        fig.legend(ncol=4, loc="lower center", fontsize=24, bbox_to_anchor=(0.5, -0.11))


plt.tight_layout()
plt.show()
```
