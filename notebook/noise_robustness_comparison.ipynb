```python
from context import qnetvo as qnet
from context import utilities 

import matplotlib.pyplot as plt
from pennylane import numpy as np
```

```python
num_samples = 21
```

# Loading Phase Damping Data

## Chain Uniform Phase Damping Data

```python
chain_uniform_pd_dir = "../script/data/chain_n-local_uniform_phase_damping/"

n3_chain_uniform_pd_regexes = [
    r"max_entangled_local_rot_n-3_.*",
    r"max_entangled_arb_n-3_.*",
    r"arb_local_rot_n-3_.*",
    r"arb_arb_n-3_.*"
]

n3_chain_uniform_pd_data = [
    utilities.analyze_data_one_param_scan(
        utilities.get_data_files(chain_uniform_pd_dir, regex)
    )
    for regex in n3_chain_uniform_pd_regexes
]

max_n3_chain_uniform_pd = [
    max(map(lambda opt_data: opt_data["max_scores"][i], n3_chain_uniform_pd_data))
    for i in range(num_samples)
]

n4_chain_uniform_pd_regexes = [
    r"max_entangled_local_rot_n-4_.*",
    r"max_entangled_arb_n-4_.*",
    r"arb_local_rot_n-4_.*",
    r"arb_arb_n-4_.*"
]

n4_chain_uniform_pd_data = [
    utilities.analyze_data_one_param_scan(
        utilities.get_data_files(chain_uniform_pd_dir, regex)
    )
    for regex in n4_chain_uniform_pd_regexes
]

max_n4_chain_uniform_pd = [
    max(map(lambda opt_data: opt_data["max_scores"][i], n4_chain_uniform_pd_data))
    for i in range(num_samples)
]
```

## Star Uniform Amplitude Damping

```python
star_uniform_ad_dir = "../script/data/star_n-local_uniform_amplitude_damping/"

n3_star_uniform_ad_regexes = [
    r"max_entangled_local_rot_n-3_.*",
    r"max_entangled_ghz_rot_n-3_.*",
    r"arb_local_rot_n-3_.*",
    r"arb_ghz_rot_n-3_.*"
]

n3_star_uniform_ad_data = [
    utilities.analyze_data_one_param_scan(
        utilities.get_data_files(star_uniform_ad_dir, regex)
    )
    for regex in n3_star_uniform_ad_regexes
]

max_n3_star_uniform_ad = [
    max(map(lambda opt_data: opt_data["max_scores"][i], n3_star_uniform_ad_data))
    for i in range(num_samples)
]

n4_star_uniform_ad_regexes = [
    r"max_entangled_local_rot_n-4_.*",
    r"max_entangled_ghz_rot_n-4_.*",
    r"arb_local_rot_n-4_.*",
    r"arb_ghz_rot_n-4_.*"
]

n4_star_uniform_ad_data = [
    utilities.analyze_data_one_param_scan(
        utilities.get_data_files(star_uniform_ad_dir, regex)
    )
    for regex in n4_star_uniform_ad_regexes
]

max_n4_star_uniform_ad = [
    max(map(lambda opt_data: opt_data["max_scores"][i], n4_star_uniform_ad_data))
    for i in range(num_samples)
]
```

```python
fig, (
    (ax1, ax2), (ax3, ax4), (ax5, ax6) 
) = plt.subplots(3, 2, figsize=(20,40))

fig.suptitle("VQO of Non-n-Locality in Noisy Networks", fontsize=18)

num_samples = 21
noise_params = np.arange(0,1.001,0.05)
quantum_bound = [np.sqrt(2)] * num_samples
classical_bound = [1] * num_samples


ax1.set_title("Uniform Phase Damping", size=16)
ax1.set_ylabel("Bell Score", size=14)
ax1.set_xlabel("Noise Parameter", size=14)

ax1.plot(noise_params, quantum_bound, label="Quantum Bound")
ax1.plot(noise_params, classical_bound, label="Classical Bound")

ax1.plot(noise_params, max_n3_chain_uniform_pd,":.", label="3-Local Chain")
ax1.plot(noise_params, max_n4_chain_uniform_pd,":d", label="4-Local Chain")
ax1.plot(noise_params, max_n3_star_uniform_pd,":^", label="3-Local Star")
ax1.plot(noise_params, max_n4_star_uniform_pd, ":x",label="4-Local Star")

ax1.legend()

ax3.set_title("Uniform Amplitude Damping", size=16)
ax3.set_ylabel("Bell Score", size=14)
ax3.set_xlabel("Noise Parameter", size=14)

ax3.plot(noise_params, quantum_bound, label="Quantum Bound")
ax3.plot(noise_params, classical_bound, label="Classical Bound")

ax3.plot(noise_params, max_n3_chain_uniform_ad, ":.", label="3-Local Chain")
ax3.plot(noise_params, max_n4_chain_uniform_ad, ":d", label="4-Local Chain")
ax3.plot(noise_params, max_n3_star_uniform_ad, ":^", label="3-Local Star")
ax3.plot(noise_params, max_n4_star_uniform_ad, ":x", label="4-Local Star")

ax3.legend()

plt.show()
```
